```python
def part1(file):
    dial = 50
    counter = 0

    with open(file, 'r') as f:
        for line in f:
            dir = line[0]
            num = int(line[1:])
            if dir == 'L':
                dial = (dial - num) % 100
            else:
                dial = (dial + num) % 100
    
            if dial % 100 == 0:
                counter += 1

    return counter
```

```python
def part2(file):
    dial = 50
    counter = 0

    with open(file, 'r') as f:
        for line in f:
            dir = line[0]
            num = int(line[1:])

            if num > 100:
                counter += floor(num / 100)
                num = num % 100
            
            if dir == 'L':
                if dial <= num and dial != 0:
                    counter += 1

                dial = (dial - num) % 100
            else:
                if dial + num >= 100:
                    counter += 1
                    
                dial = (dial + num) % 100
                
    return counter
```

```python
import os
from math import floor

file = 'day01/data.txt'

first = part1(file)
print(f"Part One: {first}")

second = part2(file)
print(f"Part Two: {second}")
```
