```python
def part1(f):
    joltage = 0

    for line in f:
        linemax = max(line[:len(line)-2])
        linesec = max(line[line.index(linemax)+1:])
        joltage += int(str(linemax)+str(linesec))
            
    return joltage
```

```python
def part2(f):
    joltage = 0
    
    def inner(start_idx: int, remaining: int) -> int:
        largest = 0
        largest_idx = 0
        for i in range(start_idx, len(line) - remaining + 1):
            v = int(line[i])
            if v > largest:
                largest = v
                largest_idx = i
        if remaining > 1:
            return largest * 10**(remaining -1) + inner(largest_idx + 1, remaining - 1)
        else:
            return largest
            
    for line in f:
        line = line.strip()
        joltage += inner(0, 12)

    return joltage
```

```python
def main():
    file = 'data.txt'

    with open(file, 'r') as f:
        content = f.readlines()
        sum1 = part1(content)
        sum2 = part2(content)
    
    print(f"sum1: {sum1}, sum2: {sum2}")
```

```python
main()
```
