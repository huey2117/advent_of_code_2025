```python
def part1(file):
    fresh = 0
    raw_ranges = []
    ranges = []

    raw = file.strip().split('\n\n')
    range_list = raw[0].split('\n')
    id_list = [int(x) for x in raw[1].split('\n')]
    
    for line in range_list:
        line = [int(x) for x in line.strip().split('-')]
        raw_ranges.append((line[0], line[1]))

    for i in id_list:
        for a, b in raw_ranges:
            if a <= i <= b:
                fresh += 1
                break
                
    return fresh
```

```python
def part2(file):
    fresh = 0
    
    range_list = file.strip().split('\n\n')[0].split('\n')
    raw_ranges = list(list(map(int, line.split('-'))) for line in range_list)
    prev = -1
    for lower, upper in sorted(raw_ranges):
        if upper <= prev:
            continue            

        fresh += upper - max(lower, prev + 1) + 1
        prev = max(upper, prev)

    return fresh
```

```python
file = 'data.txt'

with open(file, 'r') as f:
    contents = f.read()

sum1 = part1(contents)
print(sum1)   

sum2 = part2(contents)
print(sum2)
```
